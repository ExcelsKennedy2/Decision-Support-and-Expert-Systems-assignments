```python
def check_loan_eligibility(credit_score, annual_income):
    if credit_score > 700 and annual_income > 50000:
        return "Eligible for loan"
    else:
        return "Not eligible for loan"

# Test Cases
applicant1 = check_loan_eligibility(750, 60000)
applicant2 = check_loan_eligibility(680, 55000)
applicant3 = check_loan_eligibility(720, 48000)

print(applicant1)  # Output: Eligible for loan
print(applicant2)  # Output: Not eligible for loan
print(applicant3)  # Output: Not eligible for loan
```

```python
def classify_bmi(weight, height):
    bmi = weight / (height ** 2)  # BMI formula: weight (kg) / height^2 (m)
    
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal weight"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# Test Cases
person1 = classify_bmi(70, 1.75)  # Normal weight
person2 = classify_bmi(50, 1.75)  # Underweight
person3 = classify_bmi(90, 1.75)  # Overweight
person4 = classify_bmi(110, 1.75) # Obese

print(person1)  # Output: Normal weight
print(person2)  # Output: Underweight
print(person3)  # Output: Overweight
print(person4)  # Output: Obese
```

```python
class CourseFrame:
    def __init__(self, course_code, course_title, instructor, credit_hours, prerequisites=[]):
        self.attributes = {
            "Course Code": course_code,
            "Course Title": course_title,
            "Instructor": instructor,
            "Credit Hours": credit_hours,
            "Prerequisites": prerequisites
        }

    def get_attribute(self, key):
        return self.attributes.get(key, "Attribute not found")

    def set_attribute(self, key, value):
        self.attributes[key] = value

    def display_info(self):
        for key, value in self.attributes.items():
            print(f"{key}: {value}")

# Example Usage
course = CourseFrame("CS401", "Artificial Intelligence", "Dr. Smith", 3, ["CS301"])
course.display_info()

# Modify an attribute
course.set_attribute("Instructor", "Dr. Jane Doe")
print("\nUpdated Course Information:")
course.display_info()
```

```python
class BookFrame:
    def __init__(self, title, author, isbn, genre, available=True):
        self.attributes = {
            "Title": title,
            "Author": author,
            "ISBN": isbn,
            "Genre": genre,
            "Available": available
        }

    def get_attribute(self, key):
        return self.attributes.get(key, "Attribute not found")

    def set_attribute(self, key, value):
        self.attributes[key] = value

    def display_info(self):
        for key, value in self.attributes.items():
            print(f"{key}: {value}")

# Example Usage
book = BookFrame("Python Programming", "John Doe", "123-4567890123", "Computer Science")
book.display_info()

# Updating book availability
book.set_attribute("Available", False)
print("\nAfter updating availability:")
book.display_info()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes (Concepts)
nodes = ["Animal", "Mammal", "Reptile", "Dog", "Cat", "Lizard", "Has Fur", "Lays Eggs"]
G.add_nodes_from(nodes)

# Add edges (Relationships)
edges = [
    ("Dog", "Mammal"), ("Cat", "Mammal"), ("Lizard", "Reptile"),
    ("Mammal", "Animal"), ("Reptile", "Animal"),
    ("Mammal", "Has Fur"), ("Reptile", "Lays Eggs")
]
G.add_edges_from(edges)

# Draw the semantic network
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2500, font_size=10)
plt.title("Semantic Network for Animal Classification")
plt.show()
```

```python
from owlready2 import *

# Create an ontology
onto = get_ontology("http://example.org/university.owl")

with onto:
    class University(Thing):
        pass

    class Faculty(Thing):
        pass

    class Department(Thing):
        pass

    class Course(Thing):
        pass

    # Define relationships
    class hasFaculty(University >> Faculty):
        pass

    class hasDepartment(Faculty >> Department):
        pass

    class offersCourse(Department >> Course):
        pass

# Create instances
uni = University("Rongo_University")
faculty_sci = Faculty("Faculty_of_Science")
dept_cs = Department("Department_of_Computer_Science")
course_ai = Course("AI_for_Peace_Studies")

# Assign relationships
uni.hasFaculty.append(faculty_sci)
faculty_sci.hasDepartment.append(dept_cs)
dept_cs.offersCourse.append(course_ai)

# Save the ontology
onto.save(file="university.owl", format="rdfxml")

print(f"University: {uni}")
print(f"Faculties: {uni.hasFaculty}")
print(f"Departments in Science Faculty: {faculty_sci.hasDepartment}")
print(f"Courses in CS Department: {dept_cs.offersCourse}")
```

```python
from rdflib import Graph, Namespace, RDF, RDFS, Literal

# Define a namespace
ex = Namespace("http://example.org/medical/")

# Create an RDF graph
g = Graph()

# Define classes
g.add((ex.Disease, RDF.type, RDFS.Class))
g.add((ex.Symptom, RDF.type, RDFS.Class))
g.add((ex.hasSymptom, RDF.type, RDF.Property))
g.add((ex.hasSymptom, RDFS.domain, ex.Disease))
g.add((ex.hasSymptom, RDFS.range, ex.Symptom))

# Add disease and symptom instances
g.add((ex.Flu, RDF.type, ex.Disease))
g.add((ex.Fever, RDF.type, ex.Symptom))
g.add((ex.Cough, RDF.type, ex.Symptom))

# Define relationships
g.add((ex.Flu, ex.hasSymptom, ex.Fever))
g.add((ex.Flu, ex.hasSymptom, ex.Cough))

# Print the RDF graph
for s, p, o in g:
    print(f"{s} {p} {o}")
```
